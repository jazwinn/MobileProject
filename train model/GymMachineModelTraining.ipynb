```python
from roboflow import Roboflow

rf = Roboflow(api_key="") # API removed for resons
project = rf.workspace("exercise-machine").project("gym-equipment-object-detection-zpbyk")
dataset = project.version(1).download("yolo26") # It will download to a folder named 'datasets'
```

```python
from ultralytics import YOLO

# Load a pre-trained model to start from
model = YOLO('yolo26n.pt')

# Train the model
results = model.train(data=f"{dataset.location}/data.yaml", epochs=50, imgsz=640)
```
